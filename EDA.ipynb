```python
long_live= data.loc[data['Surv_status']== 1]
short_live= data.loc[data['Surv_status']== 2]
plt.plot(long_live['Age'], np.zeros_like(long_live),'o')
plt.plot(short_live['Age'], np.zeros_like(short_live),'x')

plt.show()
```

```python
plt.plot(long_live['axil_node'], np.zeros_like(long_live),'ro')
plt.plot(short_live['axil_node'], np.zeros_like(short_live),'gx')

plt.show()
```

```python
plt.plot(long_live['op_year'], np.zeros_like(long_live),'ro')
plt.plot(short_live['op_year'], np.zeros_like(short_live),'gx')

plt.show()
```

```python
counts, bin_edges= np.histogram(long_live['axil_node'],bins=10,density = True)
pdf=counts/sum(counts)
print(pdf)
cdf=np.cumsum(pdf)
print(cdf)


plt.plot(bin_edges[1:],pdf, label="PDF")
plt.plot(bin_edges[1:],cdf, label= "CDF")
plt.legend()
plt.show()
```

```python
counts, bin_edges= np.histogram(short_live['axil_node'],bins=10,density = True)
pdf=counts/sum(counts)
print(pdf)
cdf=np.cumsum(pdf)
print(cdf)


plt.plot(bin_edges[1:],pdf, label="PDF")
plt.plot(bin_edges[1:],cdf, label= "CDF")
plt.legend()
plt.show()
```

```python
counts, bin_edges= np.histogram(long_live['axil_node'],bins=10,density = True)
pdf=counts/sum(counts)
print(pdf)
cdf=np.cumsum(pdf)
print(cdf)


plt.plot(bin_edges[1:],pdf, label="PDF")
plt.plot(bin_edges[1:],cdf, label= "CDF")
plt.legend()




counts, bin_edges= np.histogram(short_live['axil_node'],bins=10,density = True)
pdf=counts/sum(counts)
print(pdf)
cdf=np.cumsum(pdf)
print(cdf)


plt.plot(bin_edges[1:],pdf, label="PDF")
plt.plot(bin_edges[1:],cdf, label= "CDF")
plt.legend()
plt.show()
```

```python
print("Means")
print(np.mean(long_live["axil_node"]))
print(np.mean((short_live["axil_node"])))

print("standard diviation")
print(np.std(long_live["axil_node"]))
print(np.std(short_live["axil_node"]))
```

```python
print("median :")
print(np.median(long_live["axil_node"]))
print(np.median(short_live['axil_node']))

print("quantiles")
print(np.percentile(long_live['axil_node'],np.arange(0,100,25)))
print(np.percentile(short_live['axil_node'], np.arange(0,100,25)))


print("90th percentile")
print(np.percentile(long_live['axil_node'],90))
print(np.percentile(short_live['axil_node'],90))

from statsmodels import robust as rb
print("medain absolute deviation")
print(rb.mad(long_live['axil_node']))
print(rb.mad(short_live['axil_node']))
```
